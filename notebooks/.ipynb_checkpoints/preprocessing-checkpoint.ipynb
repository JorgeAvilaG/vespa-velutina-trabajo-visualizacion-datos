### Datetime features

En este apartado se actualiza los tipos de las columnas que contienen fechas, para que usen el formato datetime de python. También, se crean nuevas columnas que puedan ser interesantes a la hora de explorar los datos. Como son el día de la semana, mes o año, en la que ocurrió la operación.

```python
wasp_hives["date_open"] = pd.to_datetime(wasp_hives["date_open"])
wasp_hives["date_open_day"] = wasp_hives["date_open"].dt.day
wasp_hives["date_open_week"] = wasp_hives["date_open"].dt.isocalendar().week
wasp_hives["date_open_month"] = wasp_hives["date_open"].dt.month
wasp_hives["date_open_year"] = wasp_hives["date_open"].dt.year
wasp_hives["date_open_day_of_year"] = wasp_hives["date_open"].dt.dayofyear
wasp_hives["date_open_day_of_week"] = wasp_hives["date_open"].dt.dayofweek
```

```python
wasp_hives
```

```python
wasp_hives["date_close"] = pd.to_datetime(wasp_hives["date_close"], errors="coerce")
wasp_hives["date_close_day"] = wasp_hives["date_close"].dt.day
wasp_hives["date_close_week"] = wasp_hives["date_close"].dt.isocalendar().week
wasp_hives["date_close_month"] = wasp_hives["date_close"].dt.month
wasp_hives["date_close_year"] = wasp_hives["date_close"].dt.year
wasp_hives["date_close_day_of_year"] = wasp_hives["date_close"].dt.dayofyear
wasp_hives["date_close_day_of_week"] = wasp_hives["date_close"].dt.dayofweek
```

```python
wasp_hives
```

### Location
El principal problema sobre las localización es que en los datos originales se encuentran en el formato "northing", "easting", que no se corresponde con el formato internacional de longitud y latitud. Para ello se procedió a realizar la conversión de los datos.

```python
raw_coordinates = wasp_hives[["northing", "easting"]]
raw_coordinates
```

```python
raw_coordinates.insert(2,'zone', 30)
raw_coordinates
```

```python
raw_coordinates.to_csv("../data/external/beehives_coordinates_raw.csv",index=False,)
```

Se extraen los datos de las coordenadas y se añade la columna zone con valor 30, que es correspodiente a España. Los datos se guardan y la conversión se realiza en una pag web externa que esta preparada para realizar este tipo de conversiones:
https://www.engineeringtoolbox.com/utm-latitude-longitude-d_1370.html


Leemos los datos obtenidos en la conversión, ahora tenemos los valores de longitud y latitud que nos van a ser muy utiles para realizar representaciones sobre la localización.

```python
coordinates = pd.read_csv("../data/external/beehives_coordinates.csv")[1:-1]
coordinates
```

```python
coordinates.columns = [
    "northing",
    "easting",
    "zone",
    "longitude",
    "latitude"
]
```

```python
coordinates.drop(columns=["northing", "easting", "zone"], inplace=True)
```

```python
coordinates
```

En el dataframe original las localizaciones estan en formato español, se usa ',' como separador decimal. Se  pasa la coma a punto y realizamos una operación de concatenación para añadir los valores de longitud y latitud al dataframe general.

```python
wasp_hives.easting = wasp_hives.easting.astype('str')
wasp_hives.easting = wasp_hives.easting.apply(lambda x: x.replace(',','.'))
wasp_hives.easting = wasp_hives.easting.astype('float')
wasp_hives.northing = wasp_hives.northing.astype('str')
wasp_hives.northing = wasp_hives.northing.apply(lambda x: x.replace(',','.'))
wasp_hives.northing = wasp_hives.northing.astype('float')
```

```python
wasp_hives = pd.concat([wasp_hives, coordinates], axis=1)
```

```python
wasp_hives
```

### Normalize town names

Un problema importante que se encuentra al estudiar los datos es que el nombre de los municipios se puede encontrar escrito de diferentes formas en los archivos que tenemos con datos sobre el problema. Por ejemplo, puede aparecer con el nombre del pueblo en castellano y otras veces en euskera.

Se construye una función que sirva para normalizar en formato en el que están escrito los nombre.

```python
import re
import unidecode

def normalize_string(string: str):
    lowercase_string = string.lower()
    unicode_string = unidecode.unidecode(lowercase_string)
    alphanumeric_string = re.sub(r"[^a-z]", "", unicode_string)
    return alphanumeric_string
```

```python
wasp_hives["normalized_town"] = wasp_hives["town"].apply(normalize_string)
```

```python
wasp_hives[["normalized_town", "town"]]
```

Examinando los datos se descubre que hay pueblos que aparecen con nombres diferentes. Tenemos que modificarlos manualmente en estos casos:

```python
wasp_hives.loc[wasp_hives["normalized_town"] == "abantoyciervana", "normalized_town"] = "abantoyciervanaabantozierbena"
wasp_hives.loc[wasp_hives["normalized_town"] == "munitibararbatzegi", "normalized_town"] = "munitibararbatzegigerrikaitz"
wasp_hives.loc[wasp_hives["normalized_town"] == "valledetrapaga", "normalized_town"] = "valledetrapagatrapagaran"
```

### Fruit Trees and Hives

En este apartado se van a añadir al dataframe información procedente de otros archivos.

Empezamos con la información que tenemos sobre los arboles frutales de cada municipio

```python
fruit_trees = pd.read_csv("../data/raw/FRUTALES-DECLARADOS-KOPURU.csv", delimiter=";")
```

```python
fruit_trees[:20]
```

En algunos casos faltan datos asi que eliminamos esas filas:

```python
fruit_trees[fruit_trees["NOMBRE LOCALIDAD"] == '#N/D']
```

```python
fruit_trees = fruit_trees.replace('#N/D', np.nan)
fruit_trees = fruit_trees.dropna(axis="rows", how="any")
```

```python
fruit_trees[fruit_trees["NOMBRE LOCALIDAD"] == '#N/D']
```

La información que se tiene sobre los árboles frutales es muy genérica, así que vamos a intentar agrupar la información y ver cuantos  tipos de arboles hay en cada municipio. Vamos a crear un dataframe donde para cada pueblo tengamos la cuenta del numero de frutales de cada tipo. 

```python
fruit_trees["count"] = 0
fruit_trees_count = (fruit_trees.groupby(["PRODUCTODES_C", "CODIGO MUNICIPIO", "NOMBRE LOCALIDAD"]).count()).reset_index()
fruit_trees_by_town = fruit_trees_count.pivot(index = ["CODIGO MUNICIPIO", "NOMBRE LOCALIDAD"], columns =  "PRODUCTODES_C", values = "count").reset_index()
fruit_trees_by_town = fruit_trees_by_town.fillna(value = 0)
```

```python
fruit_trees_by_town
```

Una vez que tenemos esta información extraida vamos a añadirla al dataframe general.

```python
fruit_trees_by_town.columns = map(normalize_string, list(fruit_trees_by_town.columns))
```

```python
fruit_trees_by_town
```

```python
fruit_trees_by_town = fruit_trees_by_town.drop(columns=["nombrelocalidad"])
```

Aprovechamos el código postal para hacer el merge:

```python
wasp_hives = wasp_hives.merge(fruit_trees_by_town, how="left", left_on=["town_code"], right_on=["codigomunicipio"]).drop(columns=["codigomunicipio"])
```

```python
wasp_hives
```

Para los casos en los que no hay datos lo rellenamos con 0 porque no debe haber de ese tipo de cultivo en ese pueblo.

```python
wasp_hives[["aktinidiakiwi", "arandanos", "frambuesas", "frutales", "manzano", "peral", "viedodetxakoli"]] = wasp_hives[["aktinidiakiwi", "arandanos", "frambuesas", "frutales", "manzano", "peral", "viedodetxakoli"]].fillna(0).astype(int)
```

```python
wasp_hives
```

```python
beehives.columns = ["town_name", "cp", "town_code", "beehives"]
```

Aparecen datos de colmenas para cada pueblo pero no está agrupados, vamos a sumar el numero total de colmenas por pueblo y así podremos añadir el total a nuestro dataframe general.

```python
beehives = beehives.drop(columns = ["cp"]) # no se sabe bien la diferencia entre cp y town code pero town se correspode con el identificaddor de nuestro dataframe general
beehives_by_town = (beehives.groupby(["town_code"]).sum()).reset_index()
```
